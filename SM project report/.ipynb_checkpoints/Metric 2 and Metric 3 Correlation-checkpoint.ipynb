```python
from scipy.stats import spearmanr
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Read values from the project wise result 
results = pd.read_csv("Results Project Wise_1_2_3 - Sheet1.csv")

#Statement coverage test result
branch_coverage_results = results["Branch Coverage"]
#Mutation test results
mutation_test_results = results["Mutation score"]
```

```python
print(spearmanr(branch_coverage_results, mutation_test_results))
```

```python
mutation_test_results,branch_coverage_results = zip(*sorted(zip(mutation_test_results,branch_coverage_results)))
plt.plot(mutation_test_results, branch_coverage_results)
plt.ylabel("Branch coverage results")
plt.xlabel("Mutation test results")
plt.title("Branch Coverage vs Mutation Test result")
plt.show()
```
